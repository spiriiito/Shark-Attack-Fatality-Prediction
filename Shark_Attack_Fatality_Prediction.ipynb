```python
# Step 2: Data Preprocessing
```

```python
# Dropping rows where all values are NaN
df.dropna(how='all', inplace=True)

# Dropping extra unnamed and duplicate case number columns
cols_to_drop = ['Case Number.1', 'Case Number.2', 'Unnamed: 22', 'Unnamed: 23',
                'Investigator or Source', 'href', 'href formula', 'pdf', 'original order']
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')

df.drop_duplicates(inplace=True)

print("Cleaned Nulls:")
print(df.isnull().sum())

print("\n Remaining Rows:", len(df))


# Checking for null values
print("Null values per column: ")
print(df.isnull().sum())

# Checking for duplicate rows
print("\n Number of duplicate rows: ", df.duplicated().sum())
```

```python
# Cleaning the target column ('Fatal (Y/N)') and creating a binary flag
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.upper().str.strip()
df = df[df['Fatal (Y/N)'].isin(['Y', 'N'])]
df['Fatal'] = df['Fatal (Y/N)'].map({'N': 0, 'Y': 1})
```

```python
# Cleaning and filtering 'Sex' column
df['Sex'] = df['Sex'].str.upper().str.strip()
df = df[df['Sex'].isin(['M', 'F'])]
```

```python
# Cleaning 'Activity' and 'Country'
df['Activity'] = df['Activity'].astype(str).str.lower().str.strip()
df['Country'] = df['Country'].astype(str).str.lower().str.strip()

# Dropping rows where these critical fields are missing
df.dropna(subset=['Activity', 'Sex', 'Country', 'Fatal (Y/N)'], inplace=True)
```

```python
# numeric part of Age
df['Age'] = df['Age'].astype(str).str.extract(r'(\d+)')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Filling missing values with median imputation
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Cleaning 'Species'
df['Species'] = df['Species'].astype(str).str.lower().str.strip()
df['Species'] = df['Species'].replace('nan', np.nan)
```

```python
# Dropping rows with missing Case number or Year
df.dropna(subset=['Case Number', 'Year'], inplace=True)

# Filling missing Type and Area with 'unknown'
df['Type'] = df['Type'].fillna('unknown')
df['Area'] = df['Area'].fillna('unknown')
```

```python
df.drop(columns=['Name', 'Time', 'Location'], inplace=True, errors='ignore')
```

```python
print("Final Missing Values:")
print(df.isnull().sum())

print("\n Final Dataset Shape:", df.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Normalizing the 'Age' column
scaler = StandardScaler()
df['Age_scaled'] = scaler.fit_transform(df[['Age']])
```

```python
# Original vs scaled distributions visualization
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, bins=30)
plt.title("Original Age Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['Age_scaled'], kde=True, bins=30)
plt.title("Standardized Age Distribution")

plt.tight_layout()
plt.show()
```

```python
# Label encoding (0/1) for 'Sex' column
df['Sex_encoded'] = df['Sex'].map({'M': 0, 'F': 1})

# Removing rows where 'Activity' is the string 'nan'
df = df[df['Activity'] != 'nan'].copy()

# One-hot encoding for 'Type' and 'Country' 
cols_to_encode = []
if 'Type' in df.columns:
    cols_to_encode.append('Type')
if 'Country' in df.columns:
    cols_to_encode.append('Country')

if cols_to_encode:
    df = pd.get_dummies(df, columns=cols_to_encode, drop_first=True)

# One-hot encoding for top 10 activities
top_activities = df['Activity'].value_counts().nlargest(10).index
df['activity_grouped'] = df['Activity'].apply(lambda x: x if x in top_activities else 'other')
df = pd.get_dummies(df, columns=['activity_grouped'], prefix='Activity', drop_first=True)

print("Encoded Dataset Shape:", df.shape)
print("Encoded Columns Preview:")
print(df.filter(like='Activity_').head())
```

```python
# Treating outliers
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution After Outlier Removal")

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age After Outlier Removal")

plt.tight_layout()
plt.show()
```

```python
# Visualizing dataset distribution
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Numerical columns for histogram:")
print(numeric_cols)

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
# Plotting categorical feature distributions
categorical_cols = ['Activity', 'Country', 'Type', 'Species', 'Fatal (Y/N)']

for col in categorical_cols:
    if col in df.columns:
        plt.figure(figsize=(10, 5))
        top_values = df[col].value_counts().nlargest(10)
        sns.barplot(x=top_values.values, y=top_values.index)
        plt.title(f"Top 10 Most Frequent Values in '{col}'")
        plt.xlabel("Count")
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
```

```python
# Step 3 - Feature Engineering
```

```python
# 1. Engineering is_water_activity feature
water_keywords = ['swimming', 'surfing', 'diving', 'snorkeling', 'wading', 'paddling', 'floating']
df['is_water_activity'] = df['Activity'].apply(lambda x: 1 if any(k in x for k in water_keywords) else 0)

# 2. Engineering season feature
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

def get_season(month):
    if pd.isnull(month):
        return 'unknown'
    if month in [12, 1, 2]:
        return 'winter'
    elif month in [3, 4, 5]:
        return 'spring'
    elif month in [6, 7, 8]:
        return 'summer'
    else:
        return 'fall'

df['season'] = df['Date'].dt.month.apply(get_season)
df = pd.get_dummies(df, columns=['season'], drop_first=True)

# 3. Engineering age_group feature
def group_age(age):
    if age < 13:
        return 'child'
    elif age < 20:
        return 'teen'
    elif age < 60:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(group_age)
df = pd.get_dummies(df, columns=['age_group'], drop_first=True)

# 4. Engineering is_extreme_sport feature
extreme_keywords = ['surf', 'diving', 'scuba', 'kite', 'wakeboard', 'paddle', 'windsurf', 'spearfish']
df['is_extreme_sport'] = df['Activity'].apply(lambda x: 1 if any(k in str(x).lower() for k in extreme_keywords) else 0)

# Defining target
y = df['Fatal']

# Defining features
X = df.drop(columns=['Fatal', 'Fatal (Y/N)', 'Age', 'Activity', 'Country', 'Type', 'Sex', 'Species'], errors='ignore')
```

```python
# Plotting correlation heatmap
numeric_df = df.select_dtypes(include=[np.number])

corr_matrix = numeric_df.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap with Engineered Features")
plt.tight_layout()
plt.show()
```

The correlation heatmap shows that none of the numerical features have a strong linear relationship with the target variable Fatal. The highest correlation is with Year (-0.14), suggesting a slight decrease in fatal shark attacks over time. Other features like Age, Age_scaled, and Sex_encoded show negligible correlation with fatality.

As expected, Age and Age_scaled are perfectly correlated (1.00), since one is a standardized version of the other. Only one should be used in modeling to avoid redundancy.

These results indicate that fatality is likely influenced more by categorical or contextual features (e.g., activity type, season, or shark species), which will be explored further during model development.

```python
# Training a Random Forest model for feature selection
from sklearn.ensemble import RandomForestClassifier

# Checking for non-numeric columns
non_numeric_cols = X.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols)

# Dropping them for feature importance analysis
X_numeric = X.drop(columns=non_numeric_cols)

# Dropping datetime columns
X_numeric = X_numeric.select_dtypes(exclude=['datetime64[ns]', 'datetime64[ns, UTC]'])

# Initializing the model
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fitting the model on all data (only for feature selection)
rf.fit(X_numeric, y)
```

```python
importances = pd.Series(rf.feature_importances_, index=X_numeric.columns)
importances = importances.sort_values(ascending=False)
print(importances.head(10))
importances.head(10).plot(kind='barh', title='Top 10 Feature Importances')
```

After analyzing feature importances using a Random Forest model, we observed that our engineered feature is_extreme_sport ranks 4th most important in predicting fatal shark attacks. This suggests that involvement in high-risk activities such as surfing, diving, kiteboarding, or spearfishing is a significant predictor of fatal outcomes. This insight supports the idea that extreme sports increase exposure to high-risk environments, making this a valuable feature for our model.

```python
# Selecting features by cumulative importance 
cumulative_importance = importances.cumsum()

plt.figure(figsize=(8, 5))
plt.plot(cumulative_importance.values)
plt.xlabel("Number of Features")
plt.ylabel("Cumulative Importance")
plt.title("Feature Selection based on Cumulative Importance")
plt.grid(True)
plt.show()

selected_features = cumulative_importance[cumulative_importance < 0.90].index.tolist()
print(f"Selected {len(selected_features)} features")
X_selected = X_numeric[selected_features]
print(X_selected.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
# Defining models
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
xgb_model = XGBClassifier(n_estimators=100, learning_rate=0.1, eval_metric='logloss', random_state=42)

# Training models
rf_model.fit(X_train, y_train)
gb_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)
```

```python
# Function for evaluating model performance
def evaluate_model(name, model, X_test, y_test):
    print(f"\n Model: {name}")
    y_pred = model.predict(X_test)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Classification Report
    print(classification_report(y_test, y_pred))

    # Cross-Validation Accuracy
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f" {name} 5-Fold Cross-Validation Accuracy: {np.mean(scores):.4f} ± {np.std(scores):.4f}")

evaluate_model("Random Forest", rf_model, X_test, y_test)
evaluate_model("Gradient Boosting", gb_model, X_test, y_test)
evaluate_model("XGBoost", xgb_model, X_test, y_test)
```

```python
# Step 5 - Hyperparameter Tuning
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)

# Defining parameter grid to search
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

# Setting up GridSearchCV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)
```

```python
best_rf = RandomForestClassifier(
    max_depth=20,
    max_features='sqrt',
    min_samples_leaf=1,
    min_samples_split=2,
    n_estimators=200,
    random_state=42
)

best_rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_rf.predict(X_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Cross-validation on full training data
from sklearn.model_selection import cross_val_score

scores = cross_val_score(best_rf, X_train, y_train, cv=5, scoring='f1')
print(f"\nCross-Validation F1 Score: {scores.mean():.4f} ± {scores.std():.4f}")
```
